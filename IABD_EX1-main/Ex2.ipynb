```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# Dados de treinamento
calor = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)
temperatura = np.array([30, 35, 40, 45, 60, 100, 150])
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(calor, temperatura)
# Dados de teste
calor_teste =  np.array([300, 700, 500, 600]).reshape(-1, 1)
temperatura_prev =  modelo.predict(calor_teste)
```

```python
for i in range(len(calor_teste)):
 print(f"Calor: {calor_teste[i][0]}J - Temperatura prevista: {temperatura_prev[i]} c°")
plt.scatter(calor, temperatura, color='blue', label='Dados de Treinamento')
plt.plot(calor_teste,temperatura_prev, color = 'red',label='Dados de teste')
plt.xlabel('Calor em Joules')
plt.ylabel('Temperatura em graus')
plt.title('Relação entre Calor e Temperatura')
plt.legend()
plt.grid(True)
plt.show()
```
