```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# Dados de treinamento
area = np.array([50,100,120,300]).reshape(-1, 1)
valor = np.array([180.000,300.000,375.000,600.000])
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(area, valor)
# Dados de teste
area_teste =  np.array([150, 100, 350]).reshape(-1, 1)
valor_prev =  modelo.predict(area_teste)
```

```python
for i in range(len(area_teste)):
 print(f"Area: {area_teste[i][0]} M² - valor previsto: {valor_prev[i]} R$")
plt.scatter(area, valor, color='blue', label='Dados de Treinamento')
plt.plot(area_teste,valor_prev, color = 'red',label='Dados de teste')
plt.xlabel('Area em M2')
plt.ylabel('Valor em R$')
plt.title('Relação entre area e valor do terreno')
plt.legend()
plt.grid(True)
plt.show()
```
