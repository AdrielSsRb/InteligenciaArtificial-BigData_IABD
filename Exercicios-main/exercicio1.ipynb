```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
import seaborn as sns
```

```python
#dados de treinamento
notas = np.array([8, 6, 5, 9, 4, 3, 7, 2, 1, 5, 6, 9, 8])
classes = np.array(['boa', 'ruim', 'ruim', 'boa', 'péssima', 'péssima', 'boa', 'péssima', 'péssima', 'ruim', 'ruim', 'boa', 'boa'])
```

```python
#dividir dados em teste e treino
notas_treino, notas_teste, classe_treino, classe_teste = train_test_split(notas.reshape(-1,1), classes.reshape(-1,1), test_size=0.2, random_state=42) 
```

```python
modelo = DecisionTreeClassifier()
#treinando o modelo
modelo.fit(notas_treino, classe_treino)
```

```python
fig = plt.figure(figsize=(10,8)) 
tree.plot_tree(modelo, feature_names= notas.tolist(), class_names = classes.tolist(), filled=True)
```

```python
sns.barplot(x = classes, y=notas)
sns.set(font_scale = 1)
plt.title('Boxplot das notas',fontsize = 20)
plt.ylabel('Notas', fontsize = 16)
plt.xlabel('Classes', fontsize = 16)
```

```python
y_pred = modelo.predict(notas_teste)
```
